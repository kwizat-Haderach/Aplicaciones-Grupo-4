## gravedad

```python
grav=[]

for i in T_10:
  g_10=(10/100)*4*np.pi**2/i**2
  grav.append(g_10)
  print(g_10)


for i in T_20:
  g_20=(20/100)*4*np.pi**2/i**2
  print(g_20)
  grav.append(g_20)
for i in T_30:
  g_30=(30/100)*4*np.pi**2/i**2
  print(g_30)
  grav.append(g_30)


print(f'las gravedades son {grav}')
print(f'promedio de las gravedades es {np.mean(grav)}')

```

```python
std_grav=np.std(grav,ddof=1)
print(f"La desviacin estndar de la gravedad es: {std_grav:.2f}")
```

```python
inf_grav=np.mean(grav)-std_grav
sup_grav=np.mean(grav)+std_grav
print("el limite inferior de la l global media es:",inf_grav)
print("el limite superior de la l global media es:",sup_grav)
```

```python
lim_infic5=np.mean(grav)-1.86*std_grav/np.sqrt(9)
lim_supic5=np.mean(grav)+1.86*std_grav/np.sqrt(9)
print("el limite inferior del IC de la media  gravedad es:", lim_infic5)
print("el limite superior del IC de la media gravedad es:",lim_supic5)
```

```python
# Datos de entrada
n = 9
desviacion_grav = (std_grav)**2
alpha = 0.05
df = n - 1

# Valores crticos chi-cuadrado
chi2_lower5 = chi2.ppf(alpha/2, df)
chi2_upper5 = chi2.ppf(1 - alpha/2, df)

# Intervalo de confianza para la varianza
lower_bound5 = (df * desviacion_grav) / chi2_upper5
upper_bound5 = (df * desviacion_grav) / chi2_lower5

print(f"Intervalo de confianza gravedad global para la varianza al {100*(1-alpha)}%:")
print(f"({lower_bound5:.2f}, {upper_bound5:.2f})")
```

```python
t=np.abs((np.mean(grav)-9.8037)/(std_grav/np.sqrt(9)))
print(t)

if t <= 2.306:
  print("Se acepta la hiptesis nula")
else:
  print("Se rechaza la hiptesis nula")
```

```python
into_grav=np.array([])
print("Periodo para 20cm: ", T_20)


for j in grav:
  if j>inf_grav and j<sup_grav:
    print(j, "esta dentro del intervalo de confianza")
    into_grav=np.append(into_grav,j)

  else:
    print(j, "esta fuera del intervalo de confianza")

porcentaje_grav=(len(into_grav)/len(grav))*100
print(porcentaje_grav,"%", " eat dentro del intervalo")
```

```python
#baja vs media
n1, n2, n3 = len(T_10), len(T_20),len(T_30)

# Estadstico t para dos muestras independientes (varianzas iguales)
t = (t_10 - t_20) / np.sqrt((std_dev_10**2/n1) + (std_dev_20**2/n2))

# Grados de libertad (asumiendo varianzas iguales)
gl = n1 + n2 - 2

# Valor crtico para 95% de confianza (bilateral)
tcrit = stats.t.ppf(1 - 0.025, gl)

print("t calculado:", t)
print("t crtico:", tcrit)

if np.abs(t) > tcrit:
    print("Se rechaza la hiptesis nula. Las longitudes son significativamente diferentes.")
else:
    print("No se rechaza la hiptesis nula. Las longitudes no son significativamente diferentes.")
```
